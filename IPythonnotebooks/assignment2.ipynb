```python
import pandas as pd
data = pd.read_csv("data/train.csv")
data = data.ix[:500]
data['TitleNormal']= data['Title']
data.tail()
```

```python
# use this block to clean up and normalize the Titles field into a smaller subset of titles
data['TitleNormal']=data.Title.map(lambda x: 'chef' if 'chef' in x.lower() else x)
data['TitleNormal'][data['TitleNormal'].str.contains("chef|sous|Chef|Sous|CHEF")] = 'chef' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Nurse|nurse|NURSE|RGN|RMN|RNLD")] = 'nurse' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Sales|SALES|sales")] = 'sales' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Software|SOFTWARE|software|Developer|Website")] = 'software' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Engineer|Mechanical|Engineering|ENGINEER")] = 'engineer' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Sales|SALES|sales")] = 'sales' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Manager|manager|MANAGER")] = 'manager' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Recruitment|RECRUITMENT")] = 'recruiter' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Bar|Waiting|Waiter|Waitress")] = 'service' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Health\s\w|Home\s\w|Healthcare|Support\s\w|Care\s\w|Carers")] = 'health care' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Therapist|Therapy")] = 'therapist' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Social\s\w")] = 'social worker' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Teacher|teacher|Teachers")] = 'teacher' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Dentist|GP|Doctor")] = 'doctor' # Yaay - they added this as Github 620
data['TitleNormal'][data['TitleNormal'].str.contains("Psychological|psychological")] = 'psych' # Yaay - they added this as Github 620
```
