```python
import pandas as pd
import statsmodels.formula.api as sm
from sklearn import linear_model
from patsy import dmatrices, dmatrix, DesignMatrixBuilder
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.cross_validation import train_test_split
```

```python
def readfile(size):
    if 'train' in size:
        data = pd.read_csv("data/train.csv")
    elif 'train50k' in size:
        data = pd.read_csv("data/train_50k.csv")
    elif 'train100k' in size:
        data = pd.read_csv("data/train_100k.csv")
    elif 'test' in size:
        data = pd.read_csv("data/test.csv")
    data['TitleNormal']= data['Title']
    data['TitleNormal']=data['Title']
    data = data.dropna(subset=['LocationNormalized', 'Title', 'TitleNormal'], how='any')
    data.tail(50)
    return data

data = readfile('train100k')
```

```python
# use this block to clean up and normalize the Titles field into a smaller subset of titles
def cleanTitles(data):
    data['TitleNormal'][data['TitleNormal'].str.contains("chef|sous|Chef|Sous|CHEF")] = 'chef' # Yaay - they added this as Github 620
    data['TitleNormal'][data['TitleNormal'].str.contains("Nurse|nurse|NURSE|RGN|RMN|RNLD")] = 'nurse' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Sales|SALES|sales")] = 'sales' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Android|Datacenter|Datacenter|Database|Agile|Scrum|SQL|Desktop|IT|Business\s\w|Software|UX|ux|SOFTWARE|software|Developer|Website|System\s\w")] = 'software' # Yaay - they added this as Github 620
    data['TitleNormal'][data['TitleNormal'].str.contains("Engineer|Mechanical|Engineering|ENGINEER")] = 'engineer' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Sales|SALES|sales|Account\s\w|account\s\w")] = 'sales' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Manager|manager|MANAGER")] = 'manager' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Recruiter|Recruitment|RECRUITMENT|Search\s\w")] = 'recruiter' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Bar|Waiting|Waiter|Waitress|dining|Dining")] = 'service' 
    data['TitleNormal'][data['TitleNormal'].str.contains("MRI|Endoscopy|Clinic|HEALTHCARE|health\s\w|Health\s\w|Home\s\w|Healthcare|Support\s\w|Care\s\w|Carers|HomeCare")] = 'health care' # Yaay - they added this as Github 620
    data['TitleNormal'][data['TitleNormal'].str.contains("Physiotherapist|Therapist|Therapy|psych")] = 'therapist' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Social\s\w")] = 'social worker' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Finance|Collections|Controller|Financial\s\w|Accounts|Accountant|accountant")] = 'finance' 
    data['TitleNormal'][data['TitleNormal'].str.contains("CLAIM|Insurance|insurance|Policy|Claims\s\w")] = 'insurance' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Teacher|teacher|Teachers|TEACHER|TEACHERS")] = 'teacher' 
    data['TitleNormal'][data['TitleNormal'].str.contains("CNC\s\w|Asbestos|Electrician|Warehouse|Industrial|Mechanic|Lathe|Technician")] = 'industrial worker' 
    data['TitleNormal'][data['TitleNormal'].str.contains("assistant|Secretary|Assistant|reception|Reception|Minute\s\w")] = 'assistant' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Dentist|GP|Doctor|Optometrist|optometrist|General\s\w")] = 'doctor' 
    data['TitleNormal'][data['TitleNormal'].str.contains("Psychological|psychological|Psych|psych")] = 'psych' 
    #Get only top Normalized titles
    # In doing so, I am taking upto counts of 5 per title - max 25
    topTitles = set(data.TitleNormal.value_counts().index[:50])
    data['TopTitles'] = data.TitleNormal.map(lambda x: x if x in topTitles else 'other')
    data.head()
    return data

data = cleanTitles(data)
```

```python
def cleanCompanies(data):
    #Using this block to get top 20 companies
    topCompanies = data.Company.value_counts()[:50]
    data['TopCompanies'] = data.Company.map(lambda x: x if x in topCompanies else 'other')
    data.head(40)
    return data

data = cleanCompanies(data)
```

```python
data.LocationNormalized.value_counts()[:100]
data.TopCompanies.value_counts()
```

```python
# Try to categorize Title, Location, Company, 
# Create models with Title, Location, contract type
#Clean up locations
#data['LocationNormal']=data.LocationNormalized.map(lambda x: 'london' if 'london' in x.lower() else x )
#data['LocationNormal'][data['LocationNormal'].str.contains("The\s\w")]='the city'
#data['LocationNormal'][~data['LocationNormal'].str.contains("london|the\s\w")]='other'
#data['LocationNormal'] = data.LocationNormal.map(lambda x: x if ("london|the\s\w") in x.lower() else 'other')
def cleanLocations(data):
    topLocations = data.LocationNormalized.value_counts()[:50]
    data['TopLocations'] = data.LocationNormalized.map(lambda x: x if x in topLocations else 'other')
    data.head()
    return data

data = cleanLocations(data)
```

```python
#Clean up contract type column - fillNa with unknowns
def cleanContracts(data):
    data['ContractType'].fillna('unknown',inplace=True)
    data.ContractType.value_counts
    return data

data = cleanContracts(data)
```

```python
def cleanCategories(data):
    topCategories = data.Category.value_counts()[:30]
    data['TopCategories'] = data.Category.map(lambda x: x if x in topCategories else 'other')
    data.tail(50)
    return data

data = cleanCategories(data)
```

```python

#SalaryNormalized ~ LocationNormalized -  Gives R2 of 0.525
#SalaryNormalized ~ TitleNormal - Gives R2 of 0.6
#SalaryNormalized ~ TitleNormal*LocationNormal - Gives R2 of 0.906
#SalaryNormalized ~ TitleNormal+LocationNormal - Gives R2 of 0.607 - Not much better than Location or Title independently
#SalaryNormalized ~ TitleNorma*LocationNormal + ContractTime - Gives R2 of 0.917
#SalaryNormalized ~ TitleNorma*LocationNormal + ContractTime - Gives R2 of 0.917
#SalaryNormalized ~ TitleNorma*LocationNormal + ContractType - Gives R2 of 0.953
#SalaryNormalizd ~ TitleNormal*LocationNormal+ContractType+Category - gives R2 of 0.954
#SalaryNormalizd ~ TopTitles*LocationNormal + ContractType + Category:LocationNormal + TopCompanies - gives R2 of 0.866
#SalaryNormalizd ~ TopTitles*LocationNormal+Category+LocationNormal+ContractType:TopCompanies- gives R2 of 0.865
model = sm.ols("SalaryNormalized ~ TopTitles*TopLocations + ContractType + TopLocations + TopCategories:TopCompanies", data).fit()
# Giving Rsquared of .877 so trying that.

model.summary()
```

```python
# Now implement OLS, L1 and L2 models
y, X = dmatrices('SalaryNormalized ~ TopTitles:TopLocations + ContractType + TopLocations:TopCategories + TopLocations*TopCompanies', data=data,return_type='dataframe')
```

```python
def fitModel(X,y):    
    model2 = linear_model.LinearRegression()
    model2.fit(X,y)
    model2.score(X,y)
    return model2

model2 = fitModel(X,y)
model2.score(X,y)
```

```python
model2.score(X,y) #0.40 score with training 0.52 with 50k 
#model2.intercept_
```

```python
y = np.array(y).reshape(y.shape[0])
y.shape
```

```python
def linearOrRidge(X,y,cv,model):
    if 'lasso' in model:
        if 'yes' in cv:
            l1 = linear_model.LassoCV(alphas=[0.001,0.01,10], max_iter=5000) #Cross validation avoids overfitting
        else:
            l1 = linear_model.Lasso(alpha=0.001)
        l1.fit(X,y)
        return l1
    elif 'ridge' in model:
        if 'yes' in cv:
            l2 = linear_model.RidgeCV(alphas=[0.1,1,10])
        else:
            l2 = linear_model.Ridge(alpha=1000)
        l2.fit(X,y)
        return l2

```

```python
l1 = linearOrRidge(X,y,'yes','lasso')
l1.coef_
l1.score(X,y)
```

```python
mean_absolute_error(y,l1.predict(X))
```

```python
l2 = linearOrRidge(X,y,'yes','ridge')
l2.coef_
```

```python
l2.score(X,y)
mean_absolute_error(y,l2.predict(X))
```

```python
# Training and test split
def xysplit(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=0)
    model3 = linear_model.LinearRegression()
    modelfit = model3.fit(X_train, y_train)
    modelfit.score(X_train, y_train)
    
    return X_train,X_test,y_train,y_test

X_train, X_test, y_train,y_test = xysplit(X,y)
```
