```python
from sklearn.datasets import load_files
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.datasets import fetch_20newsgroups

categories = [
    'alt.atheism',
    'talk.religion.misc',
    'comp.graphics',
    'sci.space',
]

# Load the text data
print "Loading data..."
twenty_train_subset = fetch_20newsgroups(subset='train', categories=categories)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#vectorizer = TfidfVectorizer() Restrict number of words
#vectorizer = TfidfVectorizer(max_features=10)
vectorizer = TfidfVectorizer(max_features=1000,lowercase=True,norm='l1')
X_train = vectorizer.fit_transform(twenty_train_subset.data)
```

```python
print X_train.todense()
```

```python
model = KMeans(n_clusters=3)
model.fit(X_train)
```

```python
model.cluster_centers_ #Each feature center_matri
```

```python
model.predict(X_train) # Predict tells us what each point was in the training set
```
